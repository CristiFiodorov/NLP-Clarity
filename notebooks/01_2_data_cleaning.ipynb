```python
def get_oversampled_df(df, target_col):    
    label_counts = df[target_col].value_counts()
    print(label_counts)
    
    max_count = label_counts.max()

    weights = {class_label: max_count // count for class_label, count in label_counts.items()}

    extra_entries = []
    
    for class_label, weight in weights.items():
        subset = df[df[target_col] == class_label]
        
        if weight > 1:
            extra_entries.append(pd.concat([subset] * (weight - 1))) 

    df_balanced = pd.concat([df] + extra_entries).sample(frac=1, random_state=17).reset_index(drop=True)

    print()
    print(df_balanced[target_col].value_counts())

    return df_balanced
```

```python
df_filtered_oversampled = get_oversampled_df(df_train, "evasion_label")
```
