```python
def get_oversampled_df(df, label_col):
    label_counts = df[label_col].value_counts()
    print(label_counts)
    max_count = label_counts.max()

    weights = {label: max_count // count for label, count in label_counts.items()}

    extra_entries = []
    for label, weight in weights.items():
        subset = df[df[label_col] == label]
        extra_entries.append(pd.concat([subset] * int(weight)))

    df_balanced = pd.concat(extra_entries).sample(frac=1, random_state=17).reset_index(drop=True)

    print()
    print(df_balanced[label].value_counts())

    return df_balanced
```

```python
df_filtered_oversampled = get_oversampled_df(df_train, "evasion_label")
```
